```python
import pandas as pd
df=pd.read_csv("DataFiles/AppleStore.csv")
for row in df.values:
    print(row[0],row[12])
```

```python
li=[]
for row in df.values:
    li.append(row[12])
li
```

```python
my_set=set(li)
my_new_list=list(my_set)
print("list of unique numbers: ",my_new_list)
```

```python
li1=[]
count=0
for i in range(0,len(my_new_list)):
    for j in range(0,len(li)):
        if my_new_list[i]==li[j]:
            count=count+1
    print(count)
    li1.append(count)
    count=0
li1
```

```python
import matplotlib.pyplot as plt

labels = my_new_list
sizes = li1

plt.pie(sizes,  labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
 

plt.show()
```

## Problem 6:-
#### Problem Statement:-
Using the Matplotlib library, generate a Pie Chart to categorise student marks data in a CSV file into Distinction, First Class, Second Class, Third Class and Failed Categories(based on percentages). Explode the Distinction category.
* Generate a CSV containing marks of 2000 students
* From the CSV, generate the Pie Chart

```python
import random
def genmarksdata(n,filename):
    with open(filename,'w') as marksfile:
        for i in range(0,n):
            marks=random.randint(0,100)
            marksfile.write(str(marks)+ '\n')
    return   

genmarksdata(2000,'DataFiles/marksdata.csv')
```

```python
import random
import matplotlib.pyplot as plt
def readmarksfromfile(filename):
    marks_li=[]
    with open(filename,'r') as f:
        for x in f:
            marks_li.append(int(x))
    writemarksintofile('DataFiles/marksdata.csv',marks_li)
    
def writemarksintofile(filename,arr):
    dist=0
    first=0
    second=0
    third=0
    fail=0
 
    with open(filename,'w') as reportFile:
        for i in range(0,len(arr)):
            if arr[i]>80:
                dist=dist+1
            elif(arr[i]>=60 and arr[i]<80):
                first=first+1
            elif(arr[i]>=50 and arr[i]<60):
                second=second+1
            elif(arr[i]>=40 and arr[i]<50):
                third=third+1
            else:
                fail=fail+1
    labels = 'distinction', 'first class', 'second class', 'third class','fail'
    sizes=[int(dist),int(first),int(second),int(third),int(fail)]
    explode = (0.1, 0, 0, 0, 0)
    plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90)
    plt.axis('equal')  

    plt.show()
readmarksfromfile('DataFiles/marksdata.csv')
```

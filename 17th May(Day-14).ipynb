### Problem 2:-
#### Problem Statement:- 
Define a function to add new row data to a CSV file

```python
def addRowCSV(filename,rowdata):
    df=readCSV(filename)
    df.loc[len(df)]=rowdata
    return df
rowdata=['010101','statename',123,4325,4564,9084,2773,3646,36363,25252,35353]
addRowCSV('DataFiles/Income.csv',rowdata)
```

```python
def addRowCSV(filename,rowdata):
    df=readCSV(filename) # get dataframe from CSV
    df.loc[len(df)]=rowdata # add a row to the dataframe
    df.to_csv(filename,index=False) # Send the dataframe back to CSV        
    return df
rowdata=['010101','statename',123,4325,4564,9084,2773,3646,36363,25252,3535]
addRowCSV('DataFiles/Income.csv',rowdata)
```

### Problem 3:-
#### Problem Statement:- 
Define a function to update / modify a row in aCSV file

```python
def updateRowCSV(filename, rowindex, rowdata):
    df = readCSV(filename)
    df.loc[rowindex] = rowdata
    df.to_csv(filename, index=False)
    return df

rowdata = ['04000US08','Florida',123,4325,4565,9084,83123,423,987,345,765]
readCSV('DataFiles/Income.csv')
updateRowCSV('DataFiles/Income.csv', 6, rowdata)

#df = df.drop('Unnamed: 0', axis=1)
```

### Problem 4:-
#### Problem Statement:- 
Define a function to delete the unwanted row from a CSV file

```python
def deleteRowCSV(filename,rowindex):
    df=readCSV(filename)
    df=df.drop(rowindex)
    df.to_csv(filename,index=False)
    return df
```

```python
filename='DataFiles/Income.csv'
rowindex=7
deleteRowCSV(filename,rowindex)
```

### Problem 5:-
#### Problem Statement:-
Generate marks of 2000 students in a CSV file with columns 
ROLL NO. and MARKS
from 221710400001 to 221710402000

```python
import random
def genmarksdata(n,filename):
    with open(filename,'w') as marksfile:
        for i in range(0,n):
            marks=random.randint(0,100)
            marksfile.write(str(marks)+ '\n')
    return   

genmarksdata(2000,'DataFiles/Marksdata.csv')
    
```
